```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Optional: To display plots inside notebook
%matplotlib inline
# Load the dataset
df = pd.read_excel("Sample - Superstore.xls.csv.xlsx")
# Display first 5 rows
df.head()
```

```python
# Check for missing values in each column
print("Missing values per column:\n", df.isnull().sum())
```

```python
# Get data types and info
print("\nData Info:")
df.info()
```

```python
# Basic statistics of numeric columns
print("\nSummary Statistics:")
print(df.describe())
```

```python
# Total Sales
total_sales = df['Sales'].sum()
print(f"Total Sales: ${total_sales:,.2f}")

```

```python
# Total Profit
total_profit = df['Profit'].sum()
print(f"Total Profit: ${total_profit:,.2f}")
```

```python
# Number of Orders
num_orders = df['Order ID'].nunique()
print(f"Number of Unique Orders: {num_orders}")
```

```python
# Sales by Region
sales_by_region = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
print("\nSales by Region:")
print(sales_by_region)
# Plot Sales by Region
plt.figure(figsize=(8,5))
sns.barplot(x=sales_by_region.index, y=sales_by_region.values, palette='viridis')
plt.title("Sales by Region")
plt.ylabel("Sales")
plt.xlabel("Region")
plt.show()
```

```python
# Sales by Category
sales_by_category = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
print("\nSales by Category:")
print(sales_by_category)

# Plot Sales by Category
plt.figure(figsize=(6,4))
sns.barplot(x=sales_by_category.index, y=sales_by_category.values, palette='coolwarm')
plt.title("Sales by Category")
plt.ylabel("Sales")
plt.xlabel("Category")
plt.show()
```

```python
# Convert 'Order Date' to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Extract Year-Month for grouping
df['YearMonth'] = df['Order Date'].dt.to_period('M')

# Calculate monthly sales
monthly_sales = df.groupby('YearMonth')['Sales'].sum()

# Plot monthly sales trend
plt.figure(figsize=(12,6))
monthly_sales.plot(kind='line', marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Year-Month")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Average Profit by Category
avg_profit_category = df.groupby('Category')['Profit'].mean().sort_values(ascending=False)
print("Average Profit by Category:")
print(avg_profit_category)

# Plot Average Profit by Category
plt.figure(figsize=(6,4))
sns.barplot(x=avg_profit_category.index, y=avg_profit_category.values, palette='magma')
plt.title("Average Profit by Category")
plt.ylabel("Average Profit")
plt.xlabel("Category")
plt.show()
```

```python
# Average Discount by Region
avg_discount_region = df.groupby('Region')['Discount'].mean().sort_values(ascending=False)
print("\nAverage Discount by Region:")
print(avg_discount_region)

# Plot Average Discount by Region
plt.figure(figsize=(8,5))
sns.barplot(x=avg_discount_region.index, y=avg_discount_region.values, palette='cubehelix')
plt.title("Average Discount by Region")
plt.ylabel("Average Discount")
plt.xlabel("Region")
plt.show()
```
